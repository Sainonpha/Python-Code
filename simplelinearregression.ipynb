```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df['Schooling'],df['life-expect'],marker='o')
plt.show()
```

```python
X=df[['Schooling']]
Y=df[['life-expect']]
X.head()
```

```python
Y.head()
```

```python
from sklearn.linear_model import LinearRegression
slr=LinearRegression()
slr
```

```python
slr.fit(X,Y)
```

```python
Yhat=slr.predict(X)
Yhat[0:5]
```

```python
df.head()
```

```python
df.insert(4,'predict-life-expect',Yhat,True)
df.head()
```
